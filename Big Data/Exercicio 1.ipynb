Analisando duplicatas de E.id_paciente, E.id_atendimento, E.dt_coleta

```python
%%sql
SELECT
  E.id_paciente,
  E.id_atendimento,
  E.dt_coleta,
  COUNT(*) AS total_duplicatas
FROM examlabs AS E
INNER JOIN desfechos AS D
  ON D.id_paciente = E.id_paciente
  AND D.id_atendimento = E.id_atendimento
WHERE
  E.de_exame ILIKE '%corona%'OR 
  E.de_exame ILIKE '%covid%'
GROUP BY
  E.id_paciente,
  E.id_atendimento,
  E.dt_coleta
HAVING
  COUNT(*) > 1
ORDER BY
  total_duplicatas DESC
```

```python
%%sql
SELECT 
    E.id_paciente,
    E.id_atendimento,
    E.dt_coleta,
    E.de_exame,
    E.de_resultado
FROM examlabs AS E
INNER JOIN (
    SELECT
        E.id_paciente,
        E.id_atendimento,
        E.dt_coleta,
        COUNT(*) AS total_duplicatas
    FROM examlabs AS E
    INNER JOIN desfechos AS D
        ON D.id_paciente = E.id_paciente
        AND D.id_atendimento = E.id_atendimento
    WHERE
        E.de_exame ILIKE '%corona%'
        OR E.de_exame ILIKE '%covid%'
    GROUP BY
        E.id_paciente,
        E.id_atendimento,
        E.dt_coleta
    HAVING
    COUNT(*) > 1
) AS DUPLICATAS
ON 
    E.id_atendimento = DUPLICATAS.id_atendimento AND
    E.id_paciente = DUPLICATAS.id_paciente AND
    E.dt_coleta = DUPLICATAS.dt_coleta
ORDER BY
    E.id_paciente,
    E.id_atendimento,
    E.dt_coleta
```

```python
%%sql
SELECT 
    E.de_resultado,
    COUNT(*) AS total
FROM examlabs AS E
INNER JOIN (
    SELECT
        E.id_paciente,
        E.id_atendimento,
        E.dt_coleta,
        COUNT(*) AS total_duplicatas
    FROM examlabs AS E
    INNER JOIN desfechos AS D
        ON D.id_paciente = E.id_paciente
        AND D.id_atendimento = E.id_atendimento
    WHERE
        E.de_exame ILIKE '%corona%'
        OR E.de_exame ILIKE '%covid%'
    GROUP BY
        E.id_paciente,
        E.id_atendimento,
        E.dt_coleta
    HAVING
    COUNT(*) > 1
) AS DUPLICATAS
ON 
    E.id_atendimento = DUPLICATAS.id_atendimento AND
    E.id_paciente = DUPLICATAS.id_paciente AND
    E.dt_coleta = DUPLICATAS.dt_coleta
WHERE
    E.de_exame ILIKE '%corona%'
    OR E.de_exame ILIKE '%covid%'
GROUP BY
    E.de_resultado
ORDER BY
    total DESC
```

Analisando os resultados dos exames com valores numéricos

```python
%%sql
SELECT 
    E.cd_valorreferencia,
    COUNT(*) AS total
FROM examlabs AS E
INNER JOIN desfechos AS D 
    ON E.id_paciente = D.id_paciente
    AND D.id_atendimento = E.id_atendimento
WHERE
    (
        E.de_exame ILIKE '%covid%' OR
        E.de_exame ILIKE '%corona%'
    )
    AND
    (
        E.de_resultado ~ '^-?[0-9]+(,[0-9]+)?$'
    )
GROUP BY E.cd_valorreferencia
ORDER BY total DESC
```

Vamos assumir duas coisas para a classificação entre diagnóstico de exames em que os resultados são numéricos:
- Valor de referência (se houver)
- 0,8 para os cd_valorreferencia None

Analisando os principais resultados que sejam fáceis de classificar

```python
%%sql
SELECT E.* 
FROM examlabs AS E
INNER JOIN desfechos AS D 
    ON E.id_paciente = D.id_paciente
    AND D.id_atendimento = E.id_atendimento
WHERE
    (
        E.de_exame ILIKE '%covid%' OR
        E.de_exame ILIKE '%corona%'
    ) 
    AND
    (
        E.de_resultado LIKE 'NAO DETECTADO%' OR 
        E.de_resultado LIKE 'DETECTADO%' OR
        E.de_resultado LIKE 'NÃO DETECTADO%' OR
        E.de_resultado LIKE 'NÃO REAGENTE%' OR
        E.de_resultado LIKE 'REAGENTE%' OR
        E.de_resultado LIKE 'INDETECTÁVEL%' OR
        E.de_resultado LIKE 'INDETECTAVEL%' OR
        E.de_resultado LIKE 'DETECTADO%' OR
        E.de_resultado LIKE 'Ausência%' OR
        E.de_resultado LIKE 'Detectados%' OR
        E.de_resultado LIKE 'Amostra REAGENTE%' OR
        E.de_resultado LIKE 'Amostra NÃO REAGENTE%'
    )
   
```

```python
%%sql
SELECT 
    E.id_paciente,
    E.id_atendimento,
    E.dt_coleta,
    CASE
        WHEN 
        E.de_resultado LIKE 'NAO DETECTADO%' OR 
        E.de_resultado LIKE 'NÃO DETECTADO%' OR
        E.de_resultado LIKE 'NÃO REAGENTE%' OR
        E.de_resultado LIKE 'INDETECTÁVEL%' OR
        E.de_resultado LIKE 'INDETECTAVEL%' OR
        E.de_resultado LIKE 'Ausência%' OR
        E.de_resultado LIKE 'Amostra NÃO REAGENTE%'
        THEN 'N'

        WHEN
        E.de_resultado LIKE 'DETECTADO%' OR
        E.de_resultado LIKE 'REAGENTE%' OR
        E.de_resultado LIKE 'DETECTADO%' OR
        E.de_resultado LIKE 'Detectados%' OR
        E.de_resultado LIKE 'Amostra REAGENTE%'
        THEN 'P'

        -- Resultados numéricos classificados de acordo com o valor de referência do respectivo exame
        -- Casos Positivos
        WHEN
        E.de_resultado ~ '^[0-9]+(,[0-9]+)?$' AND
        (
            (
                (E.cd_valorreferencia = 'inferior a 0,8' OR E.cd_valorreferencia = 'inferior a 0,80' OR E.cd_valorreferencia IS NULL) AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) >= 0.8
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,9' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) >= 0.9
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,7' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) >= 0.7
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,6' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) >= 0.6
            )
        )
        THEN  'P'
   
        -- Casos Negativos
        WHEN
        E.de_resultado ~ '^[0-9]+(,[0-9]+)?$' AND
        (
            (
                (E.cd_valorreferencia = 'inferior a 0,8' OR E.cd_valorreferencia = 'inferior a 0,80' OR E.cd_valorreferencia IS NULL) AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) < 0.8
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,9' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) < 0.9
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,7' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) < 0.7
            ) OR
            (
                E.cd_valorreferencia = 'inferior a 0,6' AND
                CAST(REPLACE(E.de_resultado, ',', '.') AS NUMERIC) < 0.6
            )
        )
        THEN  'N'
        
        ELSE ''
    END AS classificacao

FROM examlabs AS E
INNER JOIN desfechos AS D 
    ON E.id_paciente = D.id_paciente
    AND D.id_atendimento = E.id_atendimento
WHERE
    E.de_exame ILIKE '%covid%' OR
    E.de_exame ILIKE '%corona%'
```
