# Extracting textual features (from BERT fine-tuned classifier)

```
L = []
from tqdm.notebook import tqdm
import tensorflow as tf

t.save_tokenizer('asd')
tokenizer = t.get_tokenizer('asd')

for index,row in tqdm(df_data_train.iterrows(),total=len(df_data_train)):

  input_ids = tf.constant(tokenizer.encode(row['caption']))[None, :]  # Batch size 1
  outputs = model.layers[0](input_ids)
  last_hidden_states = outputs[0]  # The last hidden-state is the first element of the output tuple
  v = np.mean(last_hidden_states.numpy()[0],axis=0)
  L.append(v)

df_data_train['text_features'] = L
df_data_train
```

# Network Regularization

## Network generation

```
import networkx as nx
 

def cos_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def construct_graph(df, df_test, labels, m=50):
    G = nx.Graph()
    for i in range(len(df.index)):
        G.add_node(df.iloc[i]['id'])
        G.nodes[df.iloc[i]['id']]['color'] = labels[df.iloc[i]['intent']]
    for i in tqdm(range(len(df.index) // 2)):
        for k in range(i + 1, len(df.index)):
            if df.iloc[i]['intent'] == df.iloc[k]['intent']:
                G.add_edge(df.iloc[i]['id'], df.iloc[k]['id'], weight=1.0)
    for i in tqdm(range(len(df_test.index))):
        G.add_node(df_test.iloc[i]['id'])
        G.nodes[df_test.iloc[i]['id']]['color'] = labels[df_test.iloc[i]['intent']]
        neighboors = []
        dists = []
        for k in range(len(df.index)): #connect n closest similarity
          dist = (np.linalg.norm(df_test.iloc[i]['text_features']-df.iloc[k]['text_features']))**2
          dists.append(dist)
          neighboors.append((df.iloc[k]['id'], dist))
        var = np.var(np.array(dists), axis=0, dtype=np.float64)
        knn = sorted(neighboors, key=lambda n: np.exp(-n[1]/var), reverse=True)[:m]
        for j in knn:
            G.add_edge(df_test.iloc[i]['id'], j[0], weight=np.exp(-j[1]/var))

    return G
```

```
labels = {"provoke":'b', "inform":'g', "advocate":'r', "entertain":'c', "expose":'m', "express":'y', "promote":'darkorange'}
G = construct_graph(df_data_train, df_data_test, labels, m=400)
```

```
import random
from tqdm.notebook import tqdm
# Use embedding propagation
'''
    * Learn the representation for both features types
    * Join then in a mlp model, input = concat(features), labels = node['label'], than get embeddings
    * Run classifier
 
'''

def get_prop(G, prop_rate):
  D = np.zeros(len(G.nodes()))
  for node, i in zip(G.nodes(), range(len(G.nodes()))):
    D[i] = G.degree[node]
  D = np.sqrt(1/D)
  
  A = np.zeros((len(G.nodes()), len(G.nodes())))
  for node, i in zip(G.nodes(), range(len(G.nodes()))):
    for neighbor, j in zip(G.nodes(), range(len(G.nodes()))):
      if neighbor in G[node]:
        A[i,j] = A[j,i] = G[node][neighbor]['weight']
  
  L = np.dot(np.dot(np.diag(D), A), np.diag(D))
  P = np.linalg.inv(np.eye(len(G.nodes())) - prop_rate*L)

  for node, i in zip(G.nodes(), range(len(G.nodes()))):
    for neighbor, j in zip(G.nodes(), range(len(G.nodes()))):
      if neighbor in G[node]:
        G[node][neighbor]['propagation'] = P[i,j]

def regularization(G, converge=5, m=50, prop_rate=0.1, max_iter=200):
  
  nodes = []
  for node in G.nodes():
    G.nodes[node]['f_txt'] = G.nodes[node]['y_txt']*1.0
    nodes.append(node)
  
  get_prop(G, prop_rate)
  loss = 1 + converge
  old_loss = 0.0
  iter = 0
  while np.abs(loss - old_loss) > converge and iter < max_iter:
    iter += 1
    random.shuffle(nodes)
    old_loss = loss
    loss = 0.0
    for node in nodes:
      
      f_new_txt = G.nodes[node]['f_txt']*1.0
      f_old_txt = f_new_txt*1.0
      sum_w = 1.0

      # Sampling from edges 
      k = m if len(G[node].keys()) > m else len(G[node].keys())
      neighbors = random.sample(G[node].keys(), k)
      
      for neighbor in neighbors:
        w = G[node][neighbor]['propagation']
        f_new_txt += w*G.nodes[neighbor]['f_txt'] # closed form: F* = (I-alpha*L)^-1 * Y
        sum_w += w

      f_new_txt = f_new_txt/sum_w
      G.nodes[node]['f_txt'] = 1.0*f_new_txt
      new_dist_txt = np.linalg.norm(G.nodes[node]['f_txt']-f_old_txt)

      # margin ranking loss: minimize distance between new embedding and neighbors embedding
      for neighbor in neighbors:
        curr_loss_txt = np.absolute(new_dist_txt-np.linalg.norm(G.nodes[neighbor]['f_txt']-G.nodes[node]['f_txt']))
        loss += curr_loss_txt

    print(loss)
```

```
random.seed(None)
regularization(G, converge=-0.001, m=200, prop_rate=0.5, max_iter=3)
```
